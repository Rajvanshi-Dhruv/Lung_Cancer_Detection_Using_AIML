```python
import os
import random
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf

from tensorflow.keras.applications import VGG16
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc

# ------------ Setup ------------
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
from tensorflow.keras.mixed_precision import set_global_policy
set_global_policy('mixed_float16')
tf.config.optimizer.set_jit(True)

# ------------ Load and Preprocess Data ------------
data_dir = r'C:\Users\dhruv\OneDrive\Desktop\AIML\archive (1)\The IQ-OTHNCCD lung cancer dataset\colour_data'  # Update path
img_height, img_width = 160, 160
class_names = ['Normal cases', 'Bengin cases', 'Malignant cases']

images, labels = [], []

for label, class_name in enumerate(class_names):
    class_path = os.path.join(data_dir, class_name)
    if not os.path.isdir(class_path):
        continue
    for img_name in os.listdir(class_path):
        img_path = os.path.join(class_path, img_name)
        if not img_name.lower().endswith(('png', 'jpg', 'jpeg')):
            continue
        img = image.load_img(img_path, color_mode='grayscale')
        img = image.img_to_array(img)
        img_rgb = np.stack((img.squeeze(),) * 3, axis=-1)
        img_rgb = cv2.resize(img_rgb, (img_height, img_width))
        images.append(img_rgb)
        labels.append(label)

print(f'Total images loaded: {len(images)}')
images = np.array(images) / 255.0
tf.keras.backend.clear_session()
labels = to_categorical(labels, num_classes=len(class_names))

# ------------ Split Data ------------
X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.2, random_state=42)

# ------------ Create Datasets ------------
normal_train_dataset = tf.data.Dataset.from_tensor_slices((X_train, y_train)).batch(64).cache().prefetch(tf.data.AUTOTUNE)
test_dataset = tf.data.Dataset.from_tensor_slices((X_test, y_test)).batch(64).cache().prefetch(tf.data.AUTOTUNE)

# ------------ Build the Model ------------
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(img_height, img_width, 3))
base_model.trainable = False  # Freeze base model initially

model = models.Sequential([
    base_model,
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.4),
    layers.Dense(len(class_names), activation='softmax')
])

model.compile(optimizer=optimizers.SGD(learning_rate=0.001, momentum=0.9),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# ------------ Phase 1: Train Without Augmentation ------------
print("\nüöÄ Phase 1: Training on Clean Images (20 epochs)...")
history_phase1 = model.fit(
    normal_train_dataset,
    validation_data=test_dataset,
    epochs=20
)

# ------------ Phase 2: Fine-tune with Light Augmentation ------------

# Unfreeze top layers of base model
for layer in base_model.layers[-8:]:
    layer.trainable = True

# Re-compile model after unfreezing
model.compile(optimizer=optimizers.SGD(learning_rate=0.0005, momentum=0.9),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

light_augment = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.05,
    height_shift_range=0.05,
    zoom_range=0.05,
    horizontal_flip=True,
    fill_mode='nearest'
)

augmented_train_dataset = light_augment.flow(X_train, y_train, batch_size=64)

print("\nüéØ Phase 2: Fine-tuning with Light Augmentation (20 epochs)...")
history_phase2 = model.fit(
    augmented_train_dataset,
    validation_data=test_dataset,
    epochs=20
)

# ------------ Save the Model ------------
model.save('lung_cancer_vgg16_model.h5')
print("\n‚úÖ Model saved as 'lung_cancer_vgg16_model.h5'")

# ------------ Evaluate Model ------------
test_loss, test_acc = model.evaluate(test_dataset)
print(f'\nüìà Test accuracy: {test_acc * 100:.2f}%')

# ------------ Predict and Report ------------
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = np.argmax(y_test, axis=1)

print("\nClassification Report:")
print(classification_report(y_true, y_pred_classes, target_names=class_names))

cm = confusion_matrix(y_true, y_pred_classes)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# ------------ ROC Curve ------------
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(len(class_names)):
    fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_pred[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

plt.figure(figsize=(8, 6))
for i in range(len(class_names)):
    plt.plot(fpr[i], tpr[i], label=f'{class_names[i]} (AUC = {roc_auc[i]:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Each Class')
plt.legend(loc='lower right')
plt.grid()
plt.show()

# ------------ Plot Training and Validation Accuracy/Loss Graphs ------------
acc = history_phase1.history['accuracy'] + history_phase2.history['accuracy']
val_acc = history_phase1.history['val_accuracy'] + history_phase2.history['val_accuracy']
loss = history_phase1.history['loss'] + history_phase2.history['loss']
val_loss = history_phase1.history['val_loss'] + history_phase2.history['val_loss']

epochs_range = range(1, len(acc) + 1)

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.grid()

plt.show()

# ------------ Test on Random Image ------------
random_index = random.randint(0, len(X_test) - 1)
random_image = X_test[random_index]
prediction = model.predict(np.expand_dims(random_image, axis=0))
predicted_class = np.argmax(prediction)

plt.imshow(random_image)
plt.title(f'Predicted Class: {class_names[predicted_class]}')
plt.axis('off')
plt.show()
```

```python
import tkinter as tk
from tkinterdnd2 import TkinterDnD, DND_FILES
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image, ImageTk

# Load the trained model
model = load_model('lung_cancer_vgg16_model.h5')

# Class names
class_names = ['Normal cases', 'Bengin cases', 'Malignant cases']

# Function to preprocess the image and make predictions
def predict_image(img_path):
    # Load and preprocess the image
    img = image.load_img(img_path, color_mode='grayscale')  # Load as grayscale
    img = image.img_to_array(img)

    # Convert grayscale to RGB by stacking
    img_rgb = np.stack((img.squeeze(),) * 3, axis=-1)

    # Resize to match model input
    img_rgb = cv2.resize(img_rgb, (160, 160))

    # Normalize
    img_rgb = img_rgb / 255.0

    # Expand dimensions to fit model input
    img_input = np.expand_dims(img_rgb, axis=0)

    # Predict
    prediction = model.predict(img_input)
    predicted_class = np.argmax(prediction)

    return predicted_class, img_rgb

# Function to handle file drop and prediction
def on_drop(event):
    # Get the file path from the event
    img_path = event.data
    print(f"Image dropped: {img_path}")

    # Get prediction
    predicted_class, img_rgb = predict_image(img_path)
    prediction_label.config(text=f"Prediction: {class_names[predicted_class]}")

    # Display the image with prediction
    img = Image.fromarray((img_rgb * 255).astype(np.uint8))
    img = img.resize((300, 300))  # Resize for display
    img_tk = ImageTk.PhotoImage(img)
    
    # Update the image label in GUI
    image_label.config(image=img_tk)
    image_label.image = img_tk

# Set up Tkinter window with drag-and-drop
root = TkinterDnD.Tk()
root.title("Image Prediction Tool")

# Set window size
root.geometry("400x500")

# Label to display instructions
label = tk.Label(root, text="Drag and drop an image for prediction", font=('Arial', 14))
label.pack(pady=20)

# Label to display the image
image_label = tk.Label(root)
image_label.pack(pady=20)

# Label to display prediction
prediction_label = tk.Label(root, text="Prediction: None", font=('Arial', 12))
prediction_label.pack(pady=10)

# Enable drag-and-drop functionality
root.drop_target_register(DND_FILES)
root.dnd_bind('<<Drop>>', on_drop)

# Run the Tkinter event loop
root.mainloop()
```
